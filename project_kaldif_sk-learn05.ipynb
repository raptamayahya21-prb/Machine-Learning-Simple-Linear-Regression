
# Prediksi Nilai Murid

Simple Linear Regression adalah teknik statistik/machine learning paling dasar untuk memprediksi suatu nilai (output) berdasarkan satu variabel input. Hanya ada 1 fitur (X) dan 1 target (y).

## Sumber Data

Saya menemukan dataset yang cukup menarik di website Kaggle yang dapat di akses pada link berikut ini: https://www.kaggle.com/datasets/neurocipher/student-performance

Dataset tersebut adalah Performa murid pada dasarnya. data ini hanya untuk latihan prediksi dengan simple linear Regression(walaupun kurang cocok).

###Header Data

Berikut adalah header dari dataset ini

```python
import pandas as pd

path = '/content/drive/MyDrive/StudentPerformance.csv'
df = pd.read_csv(path)
df.head()
```

## Data Preprocessing

Untuk menentukan Variabel yang akan kita pilih untuk diprediksi, kita akan mencari variabel yang cukup relevan yang penting menghasilkan prediksi yang lebih akurat. Dan variabel tersebut adalah Performance Index dan Previous Scores.

###Visualisasi Data

```python
import matplotlib.pyplot as plt


plt.scatter(df['Performance Index'], df['Previous Scores'])
plt.title("Hubungan Nilai Awal dengan Nilai Akhir")
plt.xlabel('Nilai Akhir')
plt.ylabel('Nilai Awal')
plt.grid(True)
plt.show()
```

###Penyesuaian Dataset
Agar dapat di proses menggunakan algoritma ML SLR dari library sklearn, kita akan konversi dataset ke array.


```python
import numpy as np
x = np.array(df['Performance Index']).reshape(-1, 1)
y = np.array(df['Previous Scores'])

print(f'X = {x}')
print(f'Y = {y}')
```

###Training Simple Linear Regression Model
Melatih model LinearRegression dari dataset yang sudah disesuaikan menggunakan model LinearRegression dari library sklearn.linear_model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x, y)
```

###Visualisasi Simple Linear Regression Model

```python
X_vis = np.array(range(10, 100)).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(x, y)
plt.plot(X_vis, y_vis, '-r')

plt.title('Hubungan Nilai Awal dengan Nilai Akhir')
plt.xlabel('Nilai Akhir')
plt.ylabel('Nilai Awal')
plt.xlim(5,100)
plt.grid(True)
plt.show()
```

Formula Linear Regression: y=α+βx

y : response variable

x : explanatory variable

α : intercept

β : slope

```python
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
```

### Evaluasi Model dengan R² Score
R² (coefficient of determination) mengukur:

1.0 → model sangat bagus

0 → model tidak lebih baik dari “tebakan rata-rata”

negatif → model lebih buruk dari rata-rata

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, model.predict(x))
print("R² Score:", r2)
```
